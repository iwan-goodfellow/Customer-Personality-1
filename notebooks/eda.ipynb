```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from skimpy import skim
```

```python
df = pd.read_csv(r'D:\ml-baru\coba\Customer-Personality-1\data\raw\marketing_campaign.csv', delimiter="\t")
df.head()
```

```python
df.columns
```

```python
skim(df)
```

```python
for col in df.columns:
    number = df[col].nunique()
    print(f'{col} have {number} unique cols')
```

```python
# Distribusi Income
sns.histplot(df['Income'], kde=True)
plt.title('Distribusi Income')
plt.show()
```

```python
num_cols = [col for col in df.columns if df[col].dtypes != 'O' ]
num_cols
```

```python
# Korelasi antara variabel numerik
corr = df[num_cols].corr()
# Buat mask untuk segitiga bawah
mask = np.triu(np.ones_like(corr, dtype=bool))
# Plot heatmap dengan mask
plt.figure(figsize=(12, 8))
sns.heatmap(corr, mask=mask, annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi (Segitiga Bawah)')
plt.show()
```

```python
# Hubungan Education vs Income
sns.boxplot(x='Education', y='Income', data=df)
plt.title('Income berdasarkan Education')
plt.show()
```
